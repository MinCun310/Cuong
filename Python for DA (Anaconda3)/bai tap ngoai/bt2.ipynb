```python
# câu 3: Create a new column in df_final called "Country". Data in this column follows this rule:

# If the phone number starts with 84, insert "Vietnam".

# If the phone number starts with 65, insert "Singapore".

# If the phone number starts with 60, insert "Malaysia".
first_num_list = []
for phone in df_final['Phone']:
    if phone//1000000000 == 84:
        first_num_list.append('Vietnam')
    elif phone//100000000 == 65:
        first_num_list.append('Singapore')
    else:
        first_num_list.append('Malaysia')
df_final['Country'] = first_num_list
```

```python
# câu 3 cách 2
l = []
for phone in df_final['Phone']:
    if str(phone)[0:2] == '84':
        l.append('Vietnam')
    elif str(phone)[0:2] == '65':
        l.append('Singapore')
    elif str(phone)[0:2] == '60':
        l.append('Malaysia')
    else:
        l.append('')
df_final['Country'] = l
```

```python
# Câu 3 cách 3
l = []
d = {'84':'Vietnam','65':'Singapre','60':'Malaysia'}
for phone in df_final['Phone']:
    first_2 = str(phone)[0:2]
    l.append(d[first_2])
df_final['Country'] = l
```
